```python
# Selecionar sem outliers (valores superiores a R$10000)
base_sem_zero[base_sem_zero['Custo_Atendimento (R$)']<10000]
```

```python
base_sem_outliers = base_sem_zero[base_sem_zero['Custo_Atendimento (R$)']<10000]
```

```python
# Construindo um boxplot para visualizar "Custo_Atendimento (R$)" e adicionando a UF
# no eixo x.
import seaborn as  sns
import matplotlib.pyplot as plt

sns.boxplot(y = 'Custo_Atendimento (R$)',
            x = 'UF',
            data=base_sem_outliers)
plt.xticks(rotation=90)
plt.show()
```

```python
#Intervalos Interquartílicos:
Q1 = base_sem_outliers['Custo_Atendimento (R$)'].quantile(0.25)
Q3 = base_sem_outliers['Custo_Atendimento (R$)'].quantile(0.75)

#Calcular o IQR:
IQR = Q3 - Q1

#Calcular limites para a "base_sem_outliers" ("Custo_Atendimento" menor que R$10000)
limite_inf = Q1 - 1.5 * IQR
limite_sup = Q3 + 1.5 * IQR
```

```python
IQR
```

```python
limite_inf
```

```python
limite_sup
```

Excluindo-se os custos de atendimento maiores que R$10.000,00, conclui-se que o IQR (50% dos eventos) custa 670 reais, aproximadamente, podendo alcançar o custo de 2345 reais, aproximadamente.
